```python
dirName = 'tempDir2/temp2/temp'

df = pd.DataFrame(index=pd.date_range(start = dt.datetime(2020,1,1), end = dt.datetime.now(), freq='M'))

print(df)
```

```python
month_list=df.index.to_series().apply(lambda x: dt.datetime.strftime(x, '%b-%Y'))
```

```python
month_list=month_list.tolist()
```

```python
month_list
```

```python
for time in month_list:
    if os.path.isdir(gopax_base_path+time) is False:
        os.mkdir(gopax_base_path+time)
```
